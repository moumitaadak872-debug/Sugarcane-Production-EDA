SUGARCANE PRODUCTION PROJECT :

>>>This project analyzes global sugarcane production by exploring key factors such as production volume, land usage, and yield per hectare across different countries and continents. Through comprehensive Exploratory Data Analysis (EDA), the project identifies the top-producing regions, examines the relationships between these variables, and visualizes the distribution of sugarcane production. The insights gained from this analysis can support better agricultural planning and improve resource management in the sugarcane industry.

IMPORTING LIBRARIES

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

LOAD DATASET

```python
df=pd.read_csv("List of Countries by Sugarcane Production (1).csv")
```

```python
# To check first 5 rows
df.head()
```

DATA CLEANNING

```python
df["Production (Tons)"]=df["Production (Tons)"].str.replace(".","")
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].str.replace(".","").str.replace(",",".")
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].str.replace(".","") 
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].str.replace(".","").str.replace(",",".")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna().reset_index()
```

```python
df.dtypes
```

```python
df["Production (Tons)"]=df["Production (Tons)"].astype(float)
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].astype(float)
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].astype(float)
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.dtypes
```

HOW MANY COUNTRIES PRODUCE SUGARCANE FROM EACH CONTINENT?

```python
df["Continent"].value_counts().plot(kind="bar",color="seagreen")
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(df["Production (Tons)"])
plt.subplot(2,2,2)
sns.distplot(df["Production per Person (Kg)"])
plt.subplot(2,2,3)
sns.distplot(df["Acreage (Hectare)"])
plt.subplot(2,2,4)
sns.distplot(df["Yield (Kg / Hectare)"])
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
sns.boxplot(df["Production (Tons)"])
plt.subplot(2,2,2)
sns.boxplot(df["Production per Person (Kg)"])
plt.subplot(2,2,3)
sns.boxplot(df["Acreage (Hectare)"])
plt.subplot(2,2,4)
sns.boxplot(df["Yield (Kg / Hectare)"])
```

```python
df.describe()
```

WHICH COUNTRY PRODUCES MAXIMUM SUGARCANE?

```python
df_new=df[["Country","Production (Tons)"]].set_index("Country") 
df_new["Production(Tons)_percent"]=df_new["Production (Tons)"]*100/df_new["Production (Tons)"].sum()
df_new
```

```python
df_new["Production(Tons)_percent"].head(10).plot(kind="pie",autopct="%.2f")
```

```python
df_new["Production(Tons)_percent"].head(10).plot(kind="bar")
ax=sns.barplot(data=df.head(10),x="Country",y="Production (Tons)")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```
